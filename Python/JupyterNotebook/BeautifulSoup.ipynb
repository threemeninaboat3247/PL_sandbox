# Web Scripting
## 目標：GitHubのページからissue番号のリストを抽出する

```python
import requests
from bs4 import BeautifulSoup
```

```python
target_url = r"https://github.com/threemeninaboat3247/kuchinawa/issues"
```

```python
r = requests.get(target_url)  # HTML文書のgetter
soup = BeautifulSoup(r.text, "html.parser") # HTML文書のparser
```

```python
soup
```

```python
# 属性でフィルタリングして目的のDIV要素を抽出する
result = soup.find_all("div", attrs={"class":"Box-row Box-row--focus-gray p-0 mt-0 js-navigation-item js-issue-row"})
```

```python
result_all = soup.find_all("div")
```
